```python
print(" Age Distribution: Shows the distribution of customers' ages.")
print("   Inference: Most customers are aged between 25-50.")
sns.histplot(df['age'], bins=30, kde=True,  color='skyblue')
```

```python
print(" Balance Distribution: Displays the distribution of account balances.")
print("   Inference: A few customers have extremely high balances, suggesting income disparity.")
sns.histplot(df['balance'], bins=30, kde=True,color='orange')
#plt.title("Balance Distribution")
plt.show()
```

```python
print("Duration Distribution: Shows call duration distribution.")
print("   Inference: Longer call durations may indicate better engagement.")
sns.histplot(df['duration'], bins=30, kde=True, color='purple')
#axes[0, 2].set_title("Duration Distribution")
```

```python
print(" Job Distribution: Displays the count of different job categories.")
print("   Inference: Admin and blue-collar jobs are most common.")
sns.countplot(y=df['job'], order=df['job'].value_counts().index)
#axes[1, 0].set_title("Job Distribution")

```

```python
print(" Marital Status Distribution: Shows the marital status distribution.")
print("   Inference: Majority of customers are married.")
sns.countplot(x=df['marital'])
#axes[1, 1].set_title("Marital Status Distribution")
```

```python
print(" Education Distribution: Shows different education levels.")
print("   Inference: Secondary education is the most common.")
sns.countplot(x=df['education'])
#axes[1, 2].set_title("Education Distribution")
```

```python
print(" Age vs Balance Scatter Plot: Analyzes relationship between age and balance.")
print("   Inference: No strong correlation between age and balance.")
sns.scatterplot(x=df['age'], y=df['balance'], alpha=0.5, color='green')
#axes[3, 0].set_title("Age vs. Balance")
```

```python
print(" Balance vs Deposit Box Plot: Compares balance with deposit status.")
print("   Inference: Higher balances may lead to higher deposits.")
sns.boxplot(x='deposit', y='balance', data=df)
#axes[3, 1].set_title("Balance vs. Deposit")
```

```python
print(" Campaign vs Deposit Box Plot: Analyzes impact of campaigns on deposits.")
print("   Inference: Fewer calls often result in successful deposits.")
plt.figure(figsize=(5, 2))
sns.boxplot(x='deposit', y='campaign', data=df)
#plt.title("Campaign vs. Deposit")
plt.show()
```

```python
print("Correlation Heatmap: Shows correlations between numerical variables.")
print("    Inference: Duration has the highest correlation with deposits.")
plt.figure(figsize=(6, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
#plt.title("Correlation Heatmap")
plt.show()
```

```python
print(" Pairplot: Displays pairwise relationships in the dataset.")
sns.pairplot(df, hue='deposit', diag_kind='kde')
plt.show()
```
